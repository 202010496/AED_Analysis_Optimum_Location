```python
import pandas as pd
import numpy as np
import requests
import logging


version = 1
callback = "result"
apiUrl = f"https://apis.openapi.sk.com/tmap/routes?version={version}&callback={callback}"

headers = {
    "appKey": '',
    "Accept": "application/json",
    "Content-Type": "application/json; charset=UTF-8"
}

json = {
    "startX" : "126.4851928",
    "startY" : "33.4898195",

    "endX" : "126.4851928",
    "endY" : "33.4898195",
    "reqCoordType" : "WGS84GEO",
    "resCoordType" : "EPSG3857"

}

req = requests.post(apiUrl, headers=headers, json=json)
```

```python
list_dis = []
list_tim = []

for i in range(len(ori_loc['Longitude'])):
  json['startX'] = ori_loc['Longitude'][i]
  json['startY'] = ori_loc['Latitude'][i]
  for j in range(len(dict_loc['Longitude'])):
    json['endX'] = dict_loc['Longitude'][j]
    json['endY'] = dict_loc['Latitude'][j]
    req = requests.post(apiUrl, headers=headers, json=json)
    list_dis.append(req.json()['features'][0]['properties']['totalDistance'])
    list_tim.append(req.json()['features'][0]['properties']['totalTime'])


    tourspot.loc[i, 'distance'] = min(list_dis)
    tourspot.loc[i, 'time'] = min(list_tim)

  list_dis = []
  list_tim = []
```
